## 2 Sentence Splitting 
- Generally we used to understand sentence by reading words
- We will start reading sentance to understand paragraph
- So a paragraph we need to split it into a set of sentences
- So lets understand words sentences and paragraph
- This we can eadily do by using nltk package

```python
# Sentence splitting by using split() method  by using python 

myString="This is a paragraph. This is MR.Kishan It should split at the end of sentence marker ,such as a period. It can tell that the period in Mr. Daniel is not an end. Run it!, Hey How are you doing "

result=myString.split(".")

print(result)
```

```python
# Sentence splitting  by using nltk 

from nltk.tokenize import sent_tokenize 

myString="This is a paragraph. It should split at the end of sentence marker ,such as a period. It can tell that the period in Mr. Daniel is not an end. Run it!, Hey How are you doing "

tokenized_sentence=sent_tokenize(myString)

print(tokenized_sentence)
```

## 3 Tokenization 
- Tokenization refers to the process by which a large text is broken down into various pieces
- In NLP a token is the minimal piece of the text that a machine can be able to understand
- A text may be tokenized into words or sentences
  - In the case of sentence tokenization , every sentence will be identified as a token
  - In the case of word tokenization every word will be identified as a token

## 4  Word tokenization
- There are various ways through which tokenization can be done but the most popular way of doing it is through word tokenization 
- What happens in word tokenization is that a large test is broken down into words and the words are used as the tokens
  - word_tokenizer
  - sent_tokenizer
  - punkt_tokenizer
  - regexp_tokenizer
- To tokenizer a text we can still apply split() method from string 

```python
# Word tokenization by using python 

mystring="These are sentances . Let us tokenize It! Run it!"
print(mystring.split())
```

```python
# Word tokenization by using  word_tokenize function 

from nltk.tokenize import word_tokenize

mystring="These are sentances . Let us tokenize It! Run it!"

word_tokenes=word_tokenize(mystring)

print(word_tokenes)
```

```python
# Word tokenization by using  Regexp_tokenizer 

from nltk.tokenize import regexp_tokenize

mystring="These are sentances . Let us tokenize It! Run it!"

Regexp_tokenes=regexp_tokenize(mystring,pattern='\w+')

print(Regexp_tokenes)
```

```python
# Word tokenization by using  Regexp_tokenizer  # imp how to remove special symbols

from nltk.tokenize import regexp_tokenize

mystring="These are sentances . Let us 3 tokenize It! Run it!"

print(regexp_tokenize(mystring,pattern="\w+"))  # then this we should use  and w means words

```

```python
# Capture the digit from sentences  only number will show 

from nltk.tokenize import regexp_tokenize

mystring="These are 3 sentances . Let us tokenize It! Run it!"

print(regexp_tokenize(mystring,pattern="\d+"))
```

## 5 Stemming 
- This is another step in text wrangling
- From the name, you can get the meaning cutting down aa token to its root stem
- Stemming algorithm works by cutting the suffix from the word.
- In simple, it cuts either the begining or end of the word
- Consider the words "cutting"
- This word can be broken down to its root which is "cut"



```python
# Stemming

from nltk.stem import PorterStemmer

porter=PorterStemmer()

a=porter.stem("cutting")

print(a)
```

```python
from nltk.stem import PorterStemmer

e_words=["wait","waiting","waited","waits"]

ps=PorterStemmer()

for w in e_words:
    rootword=ps.stem(w)
    print(rootword)
           
```

```python

# stemming 

import nltk
from nltk.stem import PorterStemmer

porter_stemmer=PorterStemmer()
text="studied studying cries cry" 
tokenzation=nltk.word_tokenize(text)

for w in tokenzation:
    print("stemming for{} is {}".format(w,porter_stemmer.stem(w)))
```

- stemming is good for its simplicity when it comes to dealing with NLP problems
- However when more complex stemming is needed stemming is not the best option but we have lemmatization

## 6 Lemmatization
- Lemmatization is a more advanced compared to stemming
- It follow specific rules to get results
- It understand the context parts of the speech to understand the root of the word



```python
# Lemmatization 

from nltk.stem import WordNetLemmatizer

wl=WordNetLemmatizer()

print("rocks:",wl.lemmatize("rocks"))
print("corpora",wl.lemmatize("corpora"))
print("better",wl.lemmatize("better",pos='a'))
```

```python
# Lemmatization 

from nltk.stem import WordNetLemmatizer

wl = WordNetLemmatizer()

print("rocks:", wl.lemmatize("rocks"))
print("corpora:", wl.lemmatize("corpora"))
print("better:", wl.lemmatize("better", pos='a'))
```

```python
import nltk

from nltk.stem import WordNetLemmatizer

wl = WordNetLemmatizer()
text="studies studing cries cry "

tokenzation=nltk.word_tokenize(text)

for w in tokenzation:
    print("stemming for{} is {}".format(w,wl.lemmatize(w)))
```
